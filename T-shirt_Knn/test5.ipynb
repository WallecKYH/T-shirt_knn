```python
dataset=pd.read_csv("T-Shirt_Male.csv")
```

```python
dataset.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
X=dataset.iloc[:,0:2]
```

```python
Y=dataset.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.5, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import  metrics
k_range = range(1,10)
scores = {}
scores_list = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    scores[k] = metrics.accuracy_score(Y_test, Y_pred)
    scores_list.append(metrics.accuracy_score(Y_test, Y_pred))
```

```python
x=k_range
y=scores_list

ymax = max(y)
xpos = y.index(ymax)
best_k_value = x[xpos]
```

```python
#Input for weight
weight = input("Enter your weight (kg):")
#Input for height
height = input("Enter your height (cm):")
#Input for gender
gender = input("Are you male or female?")

df = pd.read_csv("T_Shirt_Female.csv") if gender == "female" else pd.read_csv("T-Shirt_Male.csv")

X_new = [[weight,height]]


def k_nearest_neighbor(df, P: list, k: int):
    def vec_abs(vec):
        return sqrt(sum([unit**2 for unit in vec]))

    def compute_vec(P_1, P_2):
        return [P_2[col] - P_1[col] for col in range(2)]

    points = []
    for row in range(df.shape[0]):
        point = [df["weightkg"][row], df["stature"][row]]
        vec = compute_vec(P, point)
        points.append((df["SizeShirt"][row] ,vec_abs(vec)))

    cluster = [size[0] for size in sorted(points, key=lambda y: y[1])[:k]]
    del points
    cluster_point = {i: cluster.count(i) for i in cluster}
    return max(cluster_point, key=cluster_point.get)



size2 = k_nearest_neighbor(df, [float(weight), float(height)], best_k_value)


print(f"Im predicting ur clothing size to : {size2}" )
```
